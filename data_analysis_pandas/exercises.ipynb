4-7. Calculate the Z-score for each numeric column of Netflix data.

$z_{score} = \frac{x - \bar{x}}{stdev} = \frac{x - \mu}{\sigma}$

```python
nflx = df[df.ticker == 'NFLX'].drop('ticker', axis=1)
```

```python
nflx.head(2)
```

```python
nflx.info()
```

```python
(nflx - nflx.mean()) / nflx.std()
```

```python
# same but through apply
nflx.apply(
    lambda x: x.sub(x.mean()).div(x.std())
)
```
