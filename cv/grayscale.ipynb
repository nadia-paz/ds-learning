Oooooops again. At least the square is red now. I need to start again and reassign R2, G2, B2 because they were assigned on BGR panel :facepalm:

```python
# let's try again
# upload
cim2 = cv2.imread('img/tulips.jpeg')
# convert from bgr to rgb
cim2 = cim2[:, :, ::-1]

# create channels
R2 = cim2[:, :, 0] # red channel
G2 = cim2[:, :, 1] # green channel
B2 = cim2[:, :, 2] # blue channel

# assign red color to the square in the middle
R2[100:400, 100: 400] = 255
G2[100:400, 100: 400] = 0
B2[100:400, 100: 400] = 0

# assign values to the image
cim2[:, :, 0] = R2
cim2[:, :, 1] = G2
cim2[:, :, 2] = B2

# plot
plt.imshow(cim2);
```

... finally :)

```python
plt.figure(1)

plt.subplot(2, 3, 1)
plt.imshow(cim2)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 3, 2)
plt.imshow(cim2)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 3, 3)
plt.imshow(cim2)
plt.xticks([])
plt.yticks([])

plt.subplot(2, 3, 4)
plt.imshow(R2, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Red', c='white')

plt.subplot(2, 3, 5)
plt.imshow(G2, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Green', c='white')

plt.subplot(2, 3, 6)
plt.imshow(B2, cmap='gray')
plt.xticks([])
plt.yticks([])
plt.title('Blue', c='white')

plt.show()
```

Create an rgb image with red, green and blue stripes

```python
# don't forget the dtype uint8, otherwise, you get warnings
rgb = np.zeros((100, 150, 3), dtype='uint8') # 100 rows, 150 columns, 3 channels
# all filled with zeros, so it's black
plt.imshow(rgb)
```

```python
# assign red to the first 50 columns
rgb[:, 0:50, 0] = 255
plt.imshow(rgb);
```

```python
rgb[:, 50:100, 1] = 255 # green channel
rgb[:, 100:150, 2] = 255 # blue channel
plt.imshow(rgb);
```
